```
#Reading two csv files at once
```

```
df = spark.read.csv(path=['dbfs:/FileStore/tables/employee1.csv', 'dbfs:/FileStore/employee2.csv'], header=True)
display(df)
df.printSchema()
```

```
from pyspark.sql.types import * 

schema = StructType().add(field='id', data_type=IntegerType())\
                    .add(field='name', data_type=StringType())

df = spark.read.csv(path=['dbfs:/FileStore/tables/employee1.csv', 'dbfs:/FileStore/employee2.csv'], schema=schema, header=True)
display(df)
df.printSchema()
```
